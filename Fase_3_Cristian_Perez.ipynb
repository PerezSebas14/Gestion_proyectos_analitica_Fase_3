```python
import pandas as pd

# Cargar los datos de los archivos CSV en dataframes
df_campanas = pd.read_csv('Anexos/Anexo 1 - Campanhas.csv', delimiter=';', encoding='latin1')
df_metricas = pd.read_csv('Anexos/Anexo 2 - Metricas.csv', delimiter=';', encoding='latin1')
df_audiencias = pd.read_csv('Anexos/Anexo 3 - Audiencias.csv', delimiter=';', encoding='latin1')

df_campanas.head()
```

```python
# Renombrar la columna 'ID_Campaña' a 'ID_Campana'
df_campanas.rename(columns={'ID_Campaña': 'ID_Campana'}, inplace=True)
df_metricas.rename(columns={'ID_Campaña': 'ID_Campana'}, inplace=True)
df_audiencias.rename(columns={'ID_Campaña': 'ID_Campana'}, inplace=True)
```

```python
# Realizar el merge de los dataframes por el ID de campaña
merged_df = df_campanas.merge(df_metricas, on='ID_Campana').merge(df_audiencias, on='ID_Campana')
```
